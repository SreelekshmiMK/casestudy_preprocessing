# 6.Do min_max scaling

```python
x=data1.drop(['Survived','Name', 'Sex' ,'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'],axis=1)
x.describe()
```

```python
data1.Survived.nunique()
```

```python
#one hot encoding
data1=pd.get_dummies(data1)
data1.head()
```

```python
#correlation
corrmatrix=data1.corr()
plt.subplots(figsize=(20,8))
sns.heatmap(corrmatrix,vmin=0.4,vmax=0.9,annot=True,linewidths=0.2)
```

```python
y=data1['Survived']
from sklearn import preprocessing
min_max=preprocessing.MinMaxScaler(feature_range=(0,1))
x=min_max.fit_transform(x)
x=pd.DataFrame(x)
x.describe()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
from sklearn import linear_model
lr=linear_model.LinearRegression()
model=lr.fit(x_train,y_train)
predictions=model.predict(x_test)
from sklearn.metrics import mean_squared_error
print('MSE is:',mean_squared_error(y_test,predictions))
from sklearn.metrics import r2_score
print('R Squarred Value is:',r2_score(y_test,predictions))
```
